```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("Line Plot")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Line Plot")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label = '^2')
plt.plot(t,t**2.2,'rs',label = '^2.2')
plt.plot(t,t**2.5,'g^',label = '^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y, linewidth = 5.0)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines,color = 'r',linewidth = 2.0)
plt.setp(lines[1], 'color','g','linewidth',2.0)
plt.grid()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
plt.figure(1)
plt.subplot(211)
plt.plot(x1,y1,'b-')
plt.subplot(212)
plt.plot(x2,y2,'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1,y1,'b-')
plt.subplot(212)
plt.plot(x2,y2,'r--')
plt.figure(2)
plt.plot(x1,y1,'g--')
plt.figure(1)
plt.subplot(211)
plt.title("Figure 1")
plt.figure(2)
plt.title("Figure 2")
plt.show()
```
